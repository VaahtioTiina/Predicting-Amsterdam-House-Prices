# Dealing With Null Values

```python
houses.isnull().sum()
```

```python
# Drop  row where price is null
houses = houses[houses.price.isnull()==False]
```

```python
# Drop houses where there is no information about the neigborhood income levels
houses = houses[houses.income_low.isnull() ==False]
```

```python
# Drop rows where there is no neighborhood safety index
houses = houses[houses.safety_index.isnull() ==False]
```

```python
houses.isnull().sum()
```

```python
# Categorical value let's check 
houses['bathrooms'] = np.where(houses['bathrooms'].isnull(), 1, houses['bathrooms'])
```

```python
# Bedrooms - we will drop this columns as we have number of rooms available
houses[houses.bedrooms.isnull()]
```

```python
houses['type'] = np.where(houses.type.isnull(), "Other", houses.type)
```

```python
houses.isnull().sum()
```

```python
houses.heating.value_counts()
```

```python
# Define missing values
houses['parking'] = np.where(houses.parking.isnull(),'Niet aanwezig', houses.parking)
```

```python
# Define missing values
houses['garage'] = np.where(houses.garage.isnull(), 'Nee', houses.garage)
```

```python
# Define missing values
houses['balcony'] = np.where(houses.balcony.isnull(), 'Niet aanwezig', houses.balcony)
houses.balcony = np.where(houses.balcony == "Niet aanwezig", 0, 1)
```

```python
houses.balcony.value_counts()
```

```python
# Define missing values
houses.garden = np. where(houses.garden.isnull(), 'Niet aanwezig', houses.garden)
houses.garden = np.where(houses.garden == "Niet aanwezig", 0, 1)
```

```python
houses.garden.value_counts()
```

```python
# Define missing values
houses.storage = np.where(houses.storage.isnull(), 'Niet aanwezig', houses.storage)
houses.storage = np.where(houses.storage == "Niet aanwezig", 0, 1)
```

```python
houses.isnull().sum()
```

```python
# Drop houses where there is no location found
houses = houses[houses['loc'] != "'-', '-''"]
```

```python
houses['loc'].value_counts()
```

```python
# Split loc to latitude and longitude
houses['lat'] = houses['loc'].apply(lambda x: x.split(",")[0])
houses['lng'] = houses['loc'].apply(lambda x: x.split(",")[1])
```

```python
houses.lat = pd.to_numeric(houses['lat'], errors='coerce')
houses.lng =pd.to_numeric(houses['lng'], errors='coerce')
```

```python
# Drop any null values
houses = houses[houses.lat.isnull()==False]
```

```python
houses.isolation.value_counts().head(15)
```

```python
houses.isolation.isnull().sum()
```

```python
# Isolation convert values
houses['isolation_floor'] = houses.isolation.apply(lambda x: 1 if re.search('vloerisola', str(x).lower()) != None else 0)
houses['isolation_double'] = houses.isolation.apply(lambda x: 1 if re.search('Dubbele beglazing', str(x)) != None else 0)
houses['isolation_part_double'] = houses.isolation.apply(lambda x: 1 if re.search('gedeeltelijk dubbele beglazing', str(x).lower()) != None else 0)
houses['isolation_roof'] = houses.isolation.apply(lambda x: 1 if re.search('dakisola', str(x).lower()) != None else 0)
houses['isolation_full'] = houses.isolation.apply(lambda x: 1 if re.search('volledig geïsoleerd', str(x).lower()) != None else 0)


                                                    
```

```python
# Drop original columns
houses.drop(columns='isolation', inplace=True)
```

```python
# Heating transform values
houses['heating_cv_ketel'] = houses.heating.apply(lambda x: 1 if re.search('cv-ketel', str(x).lower()) != None else 0)
```

```python
# Drop original columns
houses.drop(columns='heating', inplace=True)
```

```python
houses.energy_label.value_counts()
```

```python
# Assign new values for Energy labels
energy_dict = {'A': 0, 'B':1, 'C':2, 'D':3, 'E': 4, 'F':5, 'G': 6}

houses.energy_label.replace(energy_dict, inplace=True)
```

```python
#Check year values
houses.year.unique()
```

```python
# Modify values that have two years (we big the latest)
houses['age'] = houses.year.apply(lambda x: 2020 - x if len(str(x)) == 4 else 2020 - int(str(x)[4:]))

# If age is less than < 0 (house still under construction) we will give it a value 0
houses['age'] = np.where(houses['age'] < 0, 0, houses['age'])
```

```python
# Remove non-numeric characters from monthly costs
houses.monthly_cost_vve = houses.monthly_cost_vve.apply(lambda x: re.sub('[^0-9]', "", x)).astype(float)
```

```python
# For interest valculate m2 price
houses['m2 price'] = houses['price']/houses['living_area']
```

```python
# Locations - let's retrieve 4 character postcode

houses['PC4'] = houses.post_code.str[:4]
```

# Model Development

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
houses.select_dtypes(['object']).columns
```

```python
# Let's dummify the rest of the columns that we want to include in our model
houses_dummies = pd.get_dummies(data=houses,dummy_na=True,  columns=['type', 'parking', 'garage', 'PC4'], drop_first=True)
```

```python
print(houses_dummies.columns[:100])
```

```python
X = houses_dummies.drop(columns={'price', 'post_code', 'year', 'offered_since', 'status','energy_label', 'bedrooms', 'loc', 'neighborhood', 'lat', 'lng', 'm2 price'})
y = houses_dummies['price']
```

```python
# I will take a random sampel of 15 percent that will be used  as a validation set after cross validation to test the model
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size = 0.15)
```

# Random Forest

```python
# Import the model we are using and Kfold for cross-validation
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
```

```python
# Transfrom data to numpy arrays
feature_list = list(X_train.columns)
features = np.array(X_train)
target = np.array(y_train)
```

```python
# Define number of folds for cross-validation 

kf = KFold(n_splits=4,random_state=1,shuffle=True)

# Keeps track of the performance metrics during the cross-validation
accuracy_scores = [] 

for train_index, test_index in kf.split(features):

    data_train   = features[train_index]
    target_train = target[train_index]

    data_test    = features[test_index]
    target_test  = target[test_index]

    rf = RandomForestRegressor(n_estimators = 1000, 
                               random_state = 1, 
                                criterion='mse',
                               bootstrap=True)
    
    rf.fit(data_train, target_train)

    predictions = rf.predict(data_test)
    
    # Calculate absolute errors and transfrom log results back with np.exp
    errors = abs(np.exp(predictions) - np.exp(target_test))

    print('Mean Absolute Error:', (np.mean(errors)))
    
    # Calculate mean absolute percentage error
    mape = 100 * (errors / np.exp(target_test))
    
    # Accuracy based on the mape
    accuracy = 100 - np.mean(mape)
    print('Accuracy:', accuracy, '%.')

    accuracy_scores.append(accuracy)

# Final average scores
average_accuracy = np.mean(accuracy_scores)
print('Average accuracy:', average_accuracy)
```

```python
# Fir model with all data 
rf.fit(features, target)
```

# Hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#We will define a set of parameters that we can use in our grid search to find the best hyperparameters 

# Number of trees
n_estimators = [int(x) for x in np.linspace(50, 2000, 15)]

# The number of features to consider when looking for the best split:
max_features = ['auto', 'sqrt', 'log2']

# Maximum depth level of a tree
max_depth = [int(x) for x in np.linspace(10, 200, 5)]

max_depth.append(None)

# Minimum samples size to split a node
min_samples_split = [2, 4, 6, 8, 10]

# Minimum sample size  at each leaf node
min_samples_leaf = [1, 2, 3,  4]

# Bootstrap for sample selection
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Use the random grid to search for best hyperparameters
rf = RandomForestRegressor()

# Random search of parameters, using 4 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 4, verbose=2, random_state=1, n_jobs = -1)

# Fit the random search model
rf_random.fit(data_train, target_train)

print(rf_random.best_params_)
```

```python
# Model optimized with hyper parameter tuning
kf = KFold(n_splits=4,random_state=1,shuffle=True)
accuracy_scores = [] 

for train_index, test_index in kf.split(features):

    data_train   = features[train_index]
    target_train = target[train_index]

    data_test    = features[test_index]
    target_test  = target[test_index]

    rf = RandomForestRegressor(n_estimators = 885, 
                               random_state = 1,
                               min_samples_split=2,
                               min_samples_leaf=1, 
                               max_depth = 200,
                               max_features='auto',                         
                               bootstrap=True)
    
    rf.fit(data_train, target_train)

    predictions = rf.predict(data_test)
    
    errors = abs(np.exp(predictions) - np.exp(target_test))

    print('Mean Absolute Error:', (np.mean(errors)))
    
    mape = 100 * (errors / np.exp(target_test))
    
    accuracy = 100 - np.mean(mape)
    print('Accuracy:', accuracy, '%.')

    accuracy_scores.append(accuracy)

average_accuracy = np.mean(accuracy_scores)
print('Average accuracy:', average_accuracy)
```

```python
# Fit all of data
rf.fit(features, target)
predictions = rf.predict(features)
errors = abs(np.exp(predictions) - np.exp(target))
accuracy = 1 - np.mean((errors / np.exp(target)))
```

```python
plt.scatter(np.exp(predictions), np.exp(target))
print("Accuracy", accuracy)
print("R2 score", metrics.r2_score(np.exp(predictions),  np.exp(target)))
```
